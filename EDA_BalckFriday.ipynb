#### Importing Requiremnts --->

```python
import numpy as np
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Reading the data set --->

```python
data = pd.read_csv("BlackFriday.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

#### Null Values --->

```python
data.isnull().sum()
```

```python
data.drop(columns = ["Product_Category_2","Product_Category_3"], axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
for i in data:
    print(data[i].nunique(), "\t :", i, "\t:", data[i].unique())
```

#### One_hot Encodding ---> 

```python
data['OneHot_Gender'] = pd.get_dummies(data["Gender"], drop_first = True)
```

```python
del data["Gender"]
```

```python
data.head()
```

```python
data["OneHot_Gender"] = data["OneHot_Gender"].replace({False:0, True:1}).astype(int)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
count = data['OneHot_Gender'].value_counts()
count
```

#### Male to Female Purchase Ratio --->

```python
plt.pie([count[1],count[0]], labels = ["Male","Female"], autopct = "%.2f%%")
plt.show()
```

#### Ratio of Married to Un Married --->

```python
data["Marital_Status"].value_counts().plot(kind = "pie", labels = ["Un Married", "Married"], autopct = "%.2f%%")
plt.ylabel("")
```

#### Married Un Married Based on Gender --->

```python
data.groupby("OneHot_Gender")["Marital_Status"].value_counts().plot(kind = "pie", 
                                                                    labels = ["UnMa_F","Ma_F","Un_M","Ma_M"], 
                                                                    autopct = "%.2f%%")
plt.ylabel("")
```

```python
data.groupby("OneHot_Gender")["Marital_Status"].value_counts()
```

#### Gender and Marital Staus --->

```python
plt.figure(figsize = (10,7))
plt.subplot(1,2,1)
data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1].plot(kind = "pie", autopct = "%.2f%%",
                                                                       title = "Male Purchase", 
                                                                       labels = [f'Un Married {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1][0]}', f'Married {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1][1]}'], colors = ["lightskyblue","blue"])
plt.ylabel("")
plt.subplot(1,2,2)
data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[0].plot(kind = "pie", autopct = "%.2f%%",
                                                                       title = "Female Purchase", 
                                                                       labels = [f'Un Married {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[0][0]}', f'Married {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[0][1]}'], colors = ["pink","red"])
plt.ylabel("")
```
