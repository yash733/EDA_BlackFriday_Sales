#### Importing Requiremnts --->

```python
import numpy as np
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
```

#### Reading the data set --->

```python
data = pd.read_csv("BlackFriday.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

#### Null Values --->

```python
data.isnull().sum()
```

```python
data.drop(columns = ["Product_Category_2","Product_Category_3"], axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
for i in data:
    print(data[i].nunique(), "\t :", i, "\t:", data[i].unique())
```

#### One_hot Encodding ---> 

```python
data['OneHot_Gender'] = pd.get_dummies(data["Gender"], drop_first = True)
```

```python
del data["Gender"]
```

```python
data.head()
```

```python
data["OneHot_Gender"] = data["OneHot_Gender"].replace({False:0, True:1}).astype(int)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
count = data['OneHot_Gender'].value_counts()
count
```

#### Male to Female Purchase Ratio --->

```python
plt.pie([count[1],count[0]], labels = ["Male","Female"], autopct = "%.2f%%")
plt.show()
```

#### Ratio of Married to Un Married --->

```python
data["Marital_Status"].value_counts().plot(kind = "pie", labels = ["Un Married", "Married"], autopct = "%.2f%%")
plt.ylabel("")
```

```python
data.groupby("OneHot_Gender")["Marital_Status"].value_counts().plot(kind = "pie", 
                                                                    labels = ["UnMa_F","Ma_F","Un_M","Ma_M"], 
                                                                    autopct = "%.2f%%")
plt.ylabel("")
```

```python
data.groupby("OneHot_Gender")["Marital_Status"].value_counts()
```

#### Ratio of Male, Female Married and Un Married --->

```python
data.groupby("OneHot_Gender")["Marital_Status"].value_counts()[0].plot(kind = "pie", autopct = "%.2f%%", 
                                                                       title = "Female Marital Status", 
                                                                       labels = ["Un Married", "Married"], colors = ["lightblue", "purple"])
```

```python
data.groupby("OneHot_Gender")["Marital_Status"].value_counts()[1].plot(kind = "pie", autopct = "%.2f%%", 
                                                                       title = "Male Marital Status", 
                                                                       labels = ["Un Married", "Married"], colors = ["pink","orange"])
```

#### Ratio of Total Purchase based on Gender ---> 

```python
data.groupby("OneHot_Gender")["Purchase"].sum().plot(kind = "pie", 
                                                     labels = [f'Amount - {data.groupby("OneHot_Gender")["Purchase"].sum()[0]} \n Female',
                                                               f'Amount - {data.groupby("OneHot_Gender")["Purchase"].sum()[1]} \n Male'], autopct = "%.2f%%",
                                                    colors = ["lightpink", "lightblue"])
plt.ylabel("")
```

#### Purchase based on Marital Status --->

```python
data.groupby("Marital_Status")["Purchase"].sum().plot(kind = "pie", 
                                                     labels = [f'Amount - {data.groupby("OneHot_Gender")["Purchase"].sum()[0]} \n Un Married',
                                                               f'Amount - {data.groupby("OneHot_Gender")["Purchase"].sum()[1]} \n Married'], 
                                                      autopct = "%.2f%%")
plt.ylabel("")
```

```python
data.head()
```

```python
data.groupby("Marital_Status")["Purchase"].sum()
```

```python
data.groupby(data.groupby("OneHot_Gender")["Marital_Status"].value_counts()[1])["Purchase"].sum().plot(kind = "pie", autopct = "%.2f%%", 
                                                                       title = "Male Purchase", 
                                                                       labels = [f"Un Married {data.groupby(data.groupby("OneHot_Gender")["Marital_Status"].value_counts()[1])["Purchase"].sum()}", "Married"], colors = ["pink","orange"])
```

```python
data.groupby(["Marital_Status", "OneHot_Gender"])["Purchase"].sum().plot(kind = "pie", 
                                                     autopct = "%.2f%%")
plt.ylabel("")
```

```python
data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()
```

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1].plot(kind = "pie", autopct = "%.2f%%", 
                                                                           labels = [f'Amt = {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1][0]}',
                                                                                    f'Amt = {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1][1]}'],
                                                                          title = "Male")
plt.legend(labels = ["Married","Un Married"])
plt.ylabel("")

plt.subplot(1,2,2)
data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[0].plot(kind = "pie", autopct = "%.2f%%", 
                                                                           labels = [f'Amt = {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1][0]}',
                                                                                    f'Amt = {data.groupby(["OneHot_Gender","Marital_Status"])["Purchase"].sum()[1][1]}'],
                                                                          title = "Female")
```

```python
data.groupby("City_Category")["Purchase"].sum().plot(kind = "pie", labels = [f'A - {data.groupby("City_Category")["Purchase"].sum()[0]}',f'B - {data.groupby("City_Category")["Purchase"].sum()[1]}',f'C - {data.groupby("City_Category")["Purchase"].sum()[2]}']
                                                    , autopct = "%.2f%%")
plt.ylabel("")
```

```python
data.groupby(["City_Category","Stay_In_Current_City_Years"])["Purchase"].sum().plot(kind = 'bar')
```
